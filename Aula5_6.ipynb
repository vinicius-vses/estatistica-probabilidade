```python
import numpy as np
import matplotlib.pyplot as plt

populacao = np.random.uniform(0, 10, size=10000)

def gerar_amostra(n):
    return np.random.choice(populacao, size=n)

amostras = [gerar_amostra(n) for n in [10, 50, 100, 500]]

medias = [np.mean(a) for a in amostras]

plt.hist(medias[0], bins=20, alpha=0.5, label='n=10')
plt.hist(medias[1], bins=20, alpha=0.5, label='n=50')
plt.hist(medias[2], bins=20, alpha=0.5, label='n=100')
plt.hist(medias[3], bins=20, alpha=0.5, label='n=500')

plt.legend()
plt.title('Histogramas das médias de diferentes amostras')
plt.xlabel('Médias')
plt.ylabel('Frequência')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
x = np.random.normal(0, 10, 50)
y = x + np.random.normal(0, 5, 50)
slope, intercept = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, slope * x + intercept, color='red')
plt.show()
```

```python
import numpy as np


x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)

p = np.polyfit(x, y, 2)
f = np.poly1d(p)
import matplotlib.pyplot as plt


plt.scatter(x, y)
plt.plot(x, f(x), color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = x**2 + np.random.randn(100)*2

noise = np.random.randn(100)
y += noise

z = np.polyfit(x, y, 5)
p = np.poly1d(z)

corr_coef = np.corrcoef(x, y)[0, 1]

x_smooth = np.linspace(x.min(), x.max(), 100)
plt.scatter(x, y, alpha=0.5, s=10)
plt.plot(x_smooth, p(x_smooth), color='red', linewidth=2)
plt.title(f'Regressão Polinomial de grau 4\n Correlação = {corr_coef:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5

data1 = np.random.normal(mu, 1, 1000)
data2 = np.random.normal(mu, 2, 1000)
data3 = np.random.normal(mu, 3, 1000)

plt.hist(data1, bins=20, alpha=0.5, label='Desvio padrão = 1')
plt.hist(data2, bins=20, alpha=0.5, label='Desvio padrão = 2')
plt.hist(data3, bins=20, alpha=0.5, label='Desvio padrão = 3')

plt.legend(loc='upper right')
plt.title('Histogramas de distribuições normais com a mesma média')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 2

data1 = np.random.normal(0, sigma, 1000)
data2 = np.random.normal(5, sigma, 1000)
data3 = np.random.normal(10, sigma, 1000)

plt.hist(data1, bins=20, alpha=0.5, label='Média = 0')
plt.hist(data2, bins=20, alpha=0.5, label='Média = 5')
plt.hist(data3, bins=20, alpha=0.5, label='Média = 10')

plt.legend(loc='upper right')
plt.title('Histogramas de distribuições normais com o mesmo desvio padrão')

plt.show()
```
